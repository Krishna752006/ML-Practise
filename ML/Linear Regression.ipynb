```python
from sklearn.datasets import fetch_california_housing
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = fetch_california_housing()
type(df)
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset.head()
```

```python
dataset['Price'] = df.target
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_model = LinearRegression()
mse = cross_val_score(lin_model,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()

parms = {'alpha':[1e-15,1e-10,1e-8,1e-3,1,5,10,20]}

rigid_regressor = GridSearchCV(ridge,parms,scoring='neg_mean_squared_error',cv=5)
rigid_regressor.fit(X,y)
```

```python
print(rigid_regressor.best_params_)
print(rigid_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()

parms = {'alpha':[1e-15,1e-10,1e-8,1e-3,1,5,10,20]}

lasso_regressor = GridSearchCV(lasso,parms,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```
